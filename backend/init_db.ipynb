```python
from pondmemory.database.Mongo import Mongo
mongo = Mongo('localhost', 27017, 'PondMemoryDB')
```

```python
import pandas as pd
df = pd.read_csv("~/Downloads/chat_msg.csv")
df.head()
```

```python
df.shape[0]
```

```python
#需要留下Type, IsSender, CreateTime, StrContent
cnt = 0
total = df.shape[0]
for idx, row in df.iterrows():
    # print(f"idx = {idx}, Type = {row['Type']}, IsSender = {row['IsSender']}, CreateTime = {row['CreateTime']}, StrContent = {row['StrContent']}")
    obj = {
        "Type": row['Type'],
        "IsSender": row['IsSender'],
        "Timestamp": row['CreateTime'],
        "StrContent": row['StrContent']
    }
    # print(obj)
    mongo.insert_one("WechatMsg", obj)
    cnt += 1
    print(cnt/total)

```

```python
res = mongo.find_one("WechatMsg", {"Type": 1})
res
```
